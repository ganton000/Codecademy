```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#Connect to site and Scrape data.

url = "https://www.scholarships.com/financial-aid/college-scholarships/scholarships-by-type/technology-scholarships/"

page = requests.get(url)

try:
    soup = BeautifulSoup(page.text, 'html5lib')
except:
    print(page.status_code)
```

**Note**: Util_func and it's Try-Except block can be removed and the list comprehension can include an if-statement which checks whether length of list >= index_val to append the value. This is generally better practice, as except blocks with "pass" or "continue" let the code continue no matter what the error is, which could generally lead to unforeseen circumstances.

```python
#parse data and collect into lists
def util_func(lst_element, index):
    try:
        return lst_element.contents[index].strip()
    except TypeError:
        pass
    except IndexError:
        pass

deadlines = soup.select("li:nth-of-type(1)")
deadline_col = [ util_func(deadline,2) for deadline in deadlines]
deadline_col = list(filter(None, deadline_col))

amounts = soup.select("li:nth-of-type(2)")
amount_col = [ util_func(amount,2) for amount in amounts ]
amount_col = list(filter(None, amount_col))

descriptions = soup.select("li:nth-of-type(3)")
description_col = [ util_func(description,0) for description in descriptions]
description_col = list(filter(None, description_col))


#List of scholarships
scholarships = soup.select("div > h3 > a")
scholarship_col = [ scholarship.string.replace('\\', "").strip() for scholarship in scholarships ][:-2]


#check
#print(len(deadline_col), len(amount_col), len(description_col), len(scholarship_col))
```

```python
#Create dataframe
cols = ['Scholarship', 'Deadline', 'Amount', 'Description']
df_data = zip(scholarship_col, deadline_col, amount_col, description_col)
scholar_df = pd.DataFrame(columns=cols, data=df_data)

# scholar_df.set_index('Deadline',inplace=True)
# scholar_df.sort_values(by='Deadline', ascending=0)

scholar_df
```

```python
#Save to csv and excel file

scholar_df.to_csv(r"./Scholarships.csv", index=False, sep=',', encoding='utf-8')
scholar_df.to_excel("./Scholarship.xlsx")
```
